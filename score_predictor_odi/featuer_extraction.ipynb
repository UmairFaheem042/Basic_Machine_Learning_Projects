```python
cities = np.where(df['city'].isnull(), df['venue'].str.split().apply(lambda x:x[0]), df['city'])
```

```python
print(cities)
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns='venue', inplace=True)
```

```python
df.sample(5)
```

```python
df['city'].value_counts()[df['city'].value_counts() > 600]
```

```python
# consider city where count > 600 i.e. around 6 full 1st innings (12 * 50)
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
eligible_cities
```

```python
df[df['city'].isin(eligible_cities)]
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
```

```python
df.sample(5)
```

```python
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
df.sample(5)
```

```python
df['balls_bowled'] = (df['over'].astype('int') * 6) + df['ball_no'].astype('int')
```

```python
df.head()
```

```python
df['balls_left'] = 300 - df['balls_bowled']
```

```python
df.head()
```

```python
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x < 0 else x)
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x == '0' else 1)
```

```python
df.sample(5)
```

```python
# converting player_dismissed to int for calculation
df['player_dismissed'] = df['player_dismissed'].astype('int')
```

```python
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df.head(100)
```

```python
df['CRR'] = (df['current_score'] * 6) / (df['balls_bowled'])
```

```python
df
```

```python
groups = df.groupby('match_id')
```

```python
match_ids = df['match_id'].unique()
last_five = []

# for id in match_ids:
#     last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())

for id in match_ids:
    group = groups.get_group(id)
    last_five.extend(group['runs'].rolling(window=30).sum().values.tolist())
```

```python
df['runs'].dtype
```

```python
last_five
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df, on='match_id')
```

```python
completed_df = final_df[['batting_team', 'bowling_team', 'city', 'current_score', 'wickets_left', 'balls_left', 'CRR', 'last_five', 'runs_x']]
```

```python
completed_df
```

```python
completed_df.dropna(inplace=True)
```

```python
completed_df = completed_df.sample(completed_df.shape[0])
```

```python
completed_df
```

```python
X = completed_df.drop(columns='runs_x')
y = completed_df['runs_x']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city']) # Applied OneHotEncoder 
], remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', StandardScaler()),
    ('step3', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))  
])
```

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```
