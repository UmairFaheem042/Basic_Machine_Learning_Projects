```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames = []
for file in os.listdir('odis'):
    filenames.append(os.path.join('odis', file))
```

```python
final_df = pd.DataFrame()
counter = 1
df_list = [] 
for file in tqdm(filenames): # shows progress
    with open(file, 'r') as f:
        df = pd.json_normalize(safe_load(f)) # handling yaml file
        df['match_id'] = counter
        df_list.append(df)
        counter = counter + 1

final_df = pd.concat(df_list, ignore_index = True)
```

```python
count = 1
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75, 108, 150, 180, 268, 360, 443, 458, 584, 748, 982, 1052, 1111, 1226, 1345]:
        count+=1
        continue
    count+=1
    
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_dismissed = []
    teams = []
    batting_team = []
    city = []
    match_id = []
    venue = []

    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])

            try:
                player_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_dismissed.append('0')
    
    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batter': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_dismissed,
        'city': city,
        'venue': venue
    })

    delivery_df = pd.concat([delivery_df, loop_df], ignore_index=True)
```
